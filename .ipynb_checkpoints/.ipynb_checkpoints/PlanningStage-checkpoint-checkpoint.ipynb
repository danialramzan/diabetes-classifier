## **(1) Data Description:**

The Diabetes Dataset sources from the National Institute of Diabetes and Digestive and Kidney Diseases in the United States. It consists of 9 variables in total and 768 observations. The data was taken from female patients of minimum 21 years of age and of Pima Indian Heritage. Some uses of this dataset can be in cases when one wants to learn the factors which could predict the presence of Diabetes, particularly for the female population, using different statistical methods. 

| Variable | Description | Type |
|----------|-------------|------|
| 'Pregnancies' | # of pregnancies | Count |
| 'Glucose' | Glucose Level (mg/dL) | Continuous |
| 'BloodPressure' | Blood Pressure Measurement (mmHg) | Continuous |
| 'SkinThickness' | Thickness of the Skin (mm)| Continuous |
| 'Insulin' | Insulin Level in Blood | Continuous |
| 'BMI' | Body Mass Index | Continuous |
| 'DiabetesPedigreeFunction' | Diabetes Percentage (%) | Continuous |
| 'Age' | Age of the Female patient | Continuous |
| 'Outcome' | Diabetic or Not **(1 = Yes and 0 = No)** | Binary Response |

### **Visualizations**

```r
options(repr.plot.width = 17, repr.plot.height = 12) 

# GGPairs Plot
pairwise_plots <- diabetes %>%
    select(-Outcome) %>%
    ggpairs() +
    ggtitle("Pairwise Plots of the 8 Variables")

pairwise_plots
```

A GGPairs Plot is useful here as the Diabetes dataset contains only continuous variables (excluding the response). This can make it easier to explore the data for our analysis. The above pairwise plots of the 8 variables show that Glucose, BloodPressue, SkinThickness, Insulin, and BMI having values of 0 seem unreasonable. The highest correlation is 0.544 between Age and Pregnancies, but that is not too high to consider multicolinearity, therefore, multicolinearity does not seem to be an issue for our analysis as of right now. But we would have to re-visit this after dealing with the observations with 0 for Glucose, BloodPressue, SkinThickness, Insulin, and BMI. We will either drop those observations or use methods like imputation to replace 0 with the mean of that variable.

<br>

Now, I will create specific visualziations between a few variables and the response variable, 'Outcome,' which I believe will give more insights into our data and allow us to understand our data more. For example, I will analyze the following through Bar Charts and Boxplots to show differences between the groups of the response variable:
- Pregnancies and Outcome (Bar Chart)
- Age and Outcome (Boxplot)
- Glucose and Outcome (Boxplot)
- BloodPressure and Outcome (Boxplot)

```r
# Factor the binary 'Outcome' Variable
diabetes$Outcome <- factor(diabetes$Outcome, levels = c(0, 1))
```

```r
options(repr.plot.width = 15, repr.plot.height = 10) 

# Bar Chart of Pregnancies,
preg_bar_chart <- diabetes %>%
    ggplot(aes(Pregnancies, , fill = Outcome)) +
    geom_bar(position = "dodge") + 
    labs(x = "# of Pregnancies", y = "Count") + 
    ggtitle("Count of # Pregnancies for Non-Diabetic Vs Diabetic")

preg_bar_chart
```

From the bar chart, I see that non-diabetic females above the age of 21 have more pregnancies compared to diabetic patients. However, more higher number of pregnancies is unlikely for both cases. In addition, we need to remember that there is a class imbalance in this dataset as we have more non-diabetic patients than diabetic, and that could be the reason for the outcome of this distribution.

<br>

```r
options(repr.plot.width = 15, repr.plot.height = 6) 

# Box Plot of Age
age_box_plot <- diabetes %>%
    ggplot(aes(Age, Outcome, fill = Outcome)) +
    geom_boxplot() +
    labs(x = "Age", y = "Outcome") +
    ggtitle("Boxplot of Age for Diabetic Vs NonDiabetic Patients") +
    stat_summary(fun = mean, geom = "point", shape = 20, size = 5, color = "black")

age_box_plot
```

From the boxplot of age for Diabetic and Non-Diabetic patients, I see that in this dataset, more older females are diabetic. Also, there is a clear difference in the median and mean ages between both groups, with the diabetic group having a higher median and mean.

<br>

```r
options(repr.plot.width = 15, repr.plot.height = 6) 

# Box Plot of Glucose Levels
glucose_box_plot <- diabetes %>%
    ggplot(aes(Glucose, Outcome, fill = Outcome)) +
    geom_boxplot() + 
    labs(x = "Glucose Levels", y = "Outcome") +
    ggtitle("Boxplot of Glucose Levels for Diabetic Vs NonDiabetic Patients")

glucose_box_plot
```

From the above boxplot for Glucose levels, it appears that higher glucose levels may be associated with diabetic patients. Hence, glucose might be a significant predictor of diabetes for females above the age of 21.

<br>

```r
options(repr.plot.width = 12, repr.plot.height = 6) 

# Box Plot of Blood Pressure 
bp_box_plot <- diabetes %>%
    ggplot(aes(BloodPressure, Outcome, fill = Outcome)) +
    geom_boxplot() + 
    labs(x = "Blood Pressure Levels", y = "Outcome") +
    ggtitle("Boxplot of Blood Pressure Levels for Diabetic Vs NonDiabetic Patients")

bp_box_plot
```

From the above boxplot for Blood Pressure Levels, I notice that although there is a difference in median blood pressure levels for both groups, it not as big compared to the other boxplots above.

## **(4) Methods and Plan**

Re-Stating Questions of Interest:

(1) Which variables are associated with the probability of a female patient being diabetic or not?

(2) How accurately can we predict whether a female patient has diabetes or not based on the significant variables found?

For our analysis, we will begin by splitting our data into training and testing sets so we can evaluate our model in the end. 

To address our first question, the method I will use is an automated procedure, specifically, forward selection, with logistic regression since our response is a binary variable. Forward selection will allow me to find the best model out of all the possible models of different sizes based on the chosen criterion. A limitation might be that forward selection might take a long time to run but should be manageable with 8 variables.

The assumptions needed to apply Logistic Regression are to have a binary response variable, to have independent observations, and to have no multicollinearity present between the input variables.

To address our second question, we will use our final model to predict the testing set's 'Outcome' variable. Then assses our predictions using classification metrics such as accuracy, recall or precision. Accuracy may not be the best metric since we saw there is a class imblance in the response variable.
